# Try to open a csv file and put data into lines

```
try:
  fp = open('afile.csv','r')
  s = fp.read()
  fp.close()
  if '\r\n' in s:
    lines = s.split('\r\n')
  else:
    lines = s.split('\n')
except:
  print("afile.csv is not here")
```

# Print lines from the csv file

```
for line in lines:
  print(line)
```
